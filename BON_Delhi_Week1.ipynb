#### Let's examine the datatypes of all the columns

```python
areas[['Area', 'Category', 'District', 'Area_geospat']] = areas[['Area', 'Category', 'District', 'Area_geospat']].astype(str)
print(areas.dtypes)
```

#### We will convert the required string columns to string format

```python
areas[['Longitude','Latitude']] = areas.Area_geospat.str.split(",",expand=True,)
areas.head(10)
```

#### Now we will split the coordinates into two new columns Latitude and Longitude by using Pandas String split

```python
areas[['Longitude', 'Latitude']] = areas[['Longitude', 'Latitude']].astype(str)
print(areas.dtypes)
```

#### Let's convert the new columns to String type to perform string operations

```python
areas['Longitude'] = areas['Longitude'].map(lambda x: str(x).lstrip('(')).astype(str)
areas['Latitude'] = areas['Latitude'].map(lambda x: str(x).rstrip(')')).astype(str)
areas.head(10)
```

#### Now, we will strip-off the left and right braces for Longitude and Latitude respectively using the string functions

```python
areas[['Longitude', 'Latitude']] = areas[['Longitude', 'Latitude']].astype(float)
print(areas.dtypes)
```

#### Now we will convert the Latitude and Longitude to Float type to perform further operations.

```python
dl_catvs.columns = ['Category', 'Land_Cost_sq_mtr', 'Construction_Cost_sq_mtr']
print(dl_catvs.dtypes)
```

#### We will examine the data types for Category master dataframe

#### Now, we can proceed further to join both the dataframes on Category (Key) column. And to meet our data needs, inner join is okay

```python
Acost = pd.merge(areas, dl_catvs, how='inner', on = 'Category')
print(Acost.shape)
print('The dataframe has {} Districts and {} Areas.'.format(
        len(Acost['District'].unique()),
        Acost.shape[0]
    )
)
Acost.head(10)
```

#### We will calculate the total average cost for 200 sq mtrs in a given area based on the categorical cost values

```python
Acost['Cost for 200 sq mtrs'] = (Acost.Land_Cost_sq_mtr * 200) + (Acost.Construction_Cost_sq_mtr * 150)
Acost.head(10)
```
